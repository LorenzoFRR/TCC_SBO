# Etapa 2.0 — Setup do MLflow

Configura o ambiente MLflow antes do treinamento dos modelos baseline.

**Checklist:**
1. Verificar instalação do MLflow (≥ 2.x)
2. Definir URI de tracking local (`mlruns/` na raiz do projeto)
3. Criar experimento `baseline`
4. Criar estrutura de diretórios em `ARTEFATOS/ETAPA_2/`
5. Validar: UI deve mostrar experimento `baseline` vazio

## 1. Verificar instalação do MLflow

```python
import mlflow
from packaging.version import Version

print(f"MLflow version: {mlflow.__version__}")
assert Version(mlflow.__version__) >= Version("2.0"), "MLflow deve ser >= 2.x"
print("OK — versão compatível.")
```

## 2. Configurar URI de tracking e experimento `baseline`

```python
import pathlib

PROJECT_ROOT = pathlib.Path("/Users/lorenzoferreira/Documents/UFRGS/TCC_SBO")
TRACKING_URI = f"file:///{PROJECT_ROOT}/mlruns"

mlflow.set_tracking_uri(TRACKING_URI)
print(f"Tracking URI: {mlflow.get_tracking_uri()}")

# Cria o experimento se não existir; reutiliza caso já exista
experiment = mlflow.set_experiment("baseline")
print(f"Experimento ID : {experiment.experiment_id}")
print(f"Experimento    : {experiment.name}")
print(f"Artefato root  : {experiment.artifact_location}")
```

## 3. Criar estrutura de diretórios para artefatos da Etapa 2

```python
OUTPUTS = ["ET", "M_CH3OH", "x_CH3OH"]
MODELS  = ["SVR", "DT", "RF", "XGBoost", "ANN"]

ETAPA2_ROOT = PROJECT_ROOT / "ARTEFATOS" / "ETAPA_2"

for model in MODELS:
    for output in OUTPUTS:
        d = ETAPA2_ROOT / model / output
        d.mkdir(parents=True, exist_ok=True)

# Listar estrutura criada
import os
for root, dirs, files in os.walk(ETAPA2_ROOT):
    level = len(pathlib.Path(root).relative_to(ETAPA2_ROOT).parts)
    indent = "  " * level
    print(f"{indent}{pathlib.Path(root).name}/")

print("\nOK — estrutura de diretórios criada.")
```

## 4. Validação — experimento `baseline` deve estar vazio

```python
runs = mlflow.search_runs(experiment_names=["baseline"])
n_runs = len(runs)

print(f"Runs no experimento 'baseline': {n_runs}")

if n_runs == 0:
    print("OK — experimento vazio. Pronto para Etapa 2.1.")
else:
    print(f"AVISO — {n_runs} run(s) já existente(s). Verificar antes de prosseguir.")
    print(runs[["run_id", "status", "tags.mlflow.runName"]].to_string(index=False))

print(f"\nMLflow UI: mlflow ui --backend-store-uri {TRACKING_URI}")
```
